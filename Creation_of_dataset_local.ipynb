**Параметры выбора фотографий:**
1 - удалить из выборки фотографии с закрытым лицом.
2 - удалить из выборки заблюренные фотографии.
Также, я возьму bounding box height & width и уберу те, площадь которых менее 300 пикселей.
Т.к. нам нужно достаточное количество фотографицй для одного человека, то я буду собирать датасет из тех identity, для каждого из которых в исходном датасете есть как минимум 30 фотографий.

Выбранные фотографии скопирую в отдельную папку для дальнейшей работы.

```python
# уберем те строки которые заблюрены или где часть лица скрыта (в шляпе)
```

```python
attrb_good = attrb[(attrb["Blurry"] == -1) & (attrb["Wearing_Hat"] == -1)]
```

```python
attrb_good["bbox_area"] = attrb_good["width"] * attrb_good["height"]
attrb_good = attrb_good[attrb_good["bbox_area"] >= 300]
```

```python
# собираем датасет только с персонажами количество фотографий для каждого из которых больше или равно 30
```

```python
counts = attrb_good["identity"].value_counts()
valid_id = counts[counts >= 30].index
attrb_valid = attrb_good[attrb_good["identity"].isin(valid_id)]
```

```python
attrb_sample = np.random.choice(valid_id, size = 500, replace = False)
```

```python
subsample = attrb_valid[attrb_valid["identity"].isin(attrb_sample)]
subsample = subsample.groupby("identity").sample(n=30, random_state=42)
```
